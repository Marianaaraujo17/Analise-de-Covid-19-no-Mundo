```python
#Importando as bibliotecas e definindo o número máximo de colunas e linhas
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 10
pd.options.display.max_rows = 20
```

```python
#Comando para leitura do arquivo .csv
filename = 'datasets_494766_1325670_country_wise_latest.csv'
arq = pd.read_csv(filename)
```

```python
#Informações sobre o dataset
arq.info()
```

```python
#Visualização de uma parte da tabela
arq.head(24)
```

```python
#Estilos disponiveis para utilização no matplotlib 
plt.style.available
```

## Gráfico dos 10 países com maior nº de casos confirmados 

```python
plt.style.use('ggplot')
arq.groupby('Country/Region')['Confirmed'].max().sort_values().tail(10).plot(kind='barh', figsize=(8, 8))
plt.title('Dados em milhões')
```

## Gráfico dos 10 países com maior nº de mortes 

```python
plt.style.use('grayscale')
arq.groupby('Country/Region')['Deaths'].max().sort_values().tail(10).plot(kind='barh', figsize=(8, 8))
```

## Gráfico dos 10 países com menor nº de casos confirmados

```python
plt.style.use('seaborn-colorblind')
arq.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).tail(10).plot(kind='barh', figsize=(8, 8), color='green')
plt.show()
```

## Gráfico de taxa de recuperação

```python
plt.style.use('seaborn-deep')
arq.groupby('Country/Region')['Recovered / 100 Cases'].max().sort_values().tail(10).plot(kind='barh', figsize=(8, 8), color='purple')
plt.title('Taxa de recuperação a cada 100 casos')
plt.show()
```

## Gráfico de novos casos 

```python
plt.style.use('seaborn-pastel')
arq.groupby('Country/Region')['New cases'].max().sort_values().tail(5).plot(kind='pie', figsize=(8, 8))
plt.title('5 países com mais novos casos')
plt.show()
```
